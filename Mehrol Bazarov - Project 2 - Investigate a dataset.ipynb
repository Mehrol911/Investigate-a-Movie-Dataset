If you are interested in movies and want to learn interesting things about them, this dataset is a great resource to explore. There are 21 columns in this dataset. 

- `id`: Unique identifier for each movie 
- `imdb_id`: Unique identifier of each movie on IMDb 
- `popularity`: Popularity score of the movie
- `budget`: Budget of the movie in dollars 
- `revenue`: Revenue generated by the movie 
- `original_title`: Original title of the movie 
- `cast`: Name of the actors and actresses who acted in the movie
- `homepage`: Website of the movie. 
- `director`: Name of the director who directed the movie
- `tagline`: Movie’s tagline if any. 
- `keywords`: The keywords or tags related to the movie. 
- `overview`: Brief description of the movie.
- `runtime`: Duration of the movie in minutes
- `genres`: The genres of the movie.
- `production_companies`: Production companies involved in making the movie
- `release_date`: Release date of the movie 
- `vote_count`: Number of votes received by the movie
- `vote_average`: Average rating of the movie
- `release_year`: Release year of the movie
- `budget_adj`: Budget of the movie in terms of 2010 dollars, accounting for inflation
- `revenue_adj`: Revenue of the movie in terms of 2010 dollars, accounting for inflation

### Research Question 3.4  (To have more vote_counts lead to have high vote_average rating?)


```python
#scatter plot of vote_count vs vote_average using plot.scatter() function:
vote_count_average = df.plot.scatter(x='vote_count', y='vote_average', c='green', 
                                     title = "To have more vote_counts lead to have high vote_average rating?")
```

```python
#using numpy correlation between two columns which are vote_count and vote_average
vote_count = df['vote_count']
vote_average = df['vote_average']

correlation_matrix = np.corrcoef(vote_count, vote_average)
correlation_xy = correlation_matrix[0,1]
print(correlation_xy)
```

The correlation coefficient ranges between -1 and 1. A value of -1 indicates a perfect negative correlation while a value of 1 indicates a perfect positive correlation. A value of 0 indicates no correlation between the two variables.
In our case they are <b>positively correlated</b>. 

<a id='question35'></a>
### Research Question 3.5 (Is it true that movie with high vote_count tend to be popular?)


In order to get the answer for that question, let's make correlation matrix.

```python
#creating corr_matrix
corr_matrix = df.corr()

print(corr_matrix)
```

```python
# print list of tuples to see highly correlated variables
high_corr_vars = corr_matrix.where(
    (corr_matrix > 0.7) & (corr_matrix < 1.0)
).stack().index.tolist()

print(high_corr_vars)
```

As you can see from table above, the correlation between two variables are 0.802956. They are positively correlated. And answer for the question is <b>BIG YES</b>!

### Research Question 4.1 (Which movie is the most profitable and which is the opposite?)
 

```python
# Create a new column named 'profit' which will contain the difference between revenue_adj and budget_adj columns
df['profit'] = df['revenue_adj'] - df['budget_adj']
df
```

```python
#printing the maximum and minimum profit for profit column
max_profit = df['profit'].max()
min_profit = df['profit'].min()

print(f'Maximum profit: {max_profit}')
print(f'Minimum profit: {min_profit}')
```

```python
#printing title of these two films
max_profit_film = df[df['profit'] == max_profit]['title'].values[0]
min_profit_film = df[df['profit'] == min_profit]['title'].values[0]

print(f'Film with maximum profit: {max_profit_film}')
print(f'Film with minimum profit: {min_profit_film}')
```

### Research Question 4.2 (How many movies are profitable in this dataset?)

```python
# Create new column named 'profitable' which will contain 'Yes' if the value in the profit column is positive and 'No' if it is negative
df['profitable'] = np.where(df['profit'] > 0, 'Yes', 'No')
df
```
